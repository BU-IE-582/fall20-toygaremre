```python
import pandas as pd
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
Matches=pd.read_csv('/Users/toygar/Desktop/IE582/E0 2019-2020.csv',sep=str,delimiter=',')
dataframe1=pd.DataFrame(Matches)

B365H_p_home=1/dataframe1['B365H']
B365H_p_home=B365H_p_home.round(4)
dataframe1['B365H_p_home']=B365H_p_home


B365D_p_draw=1/dataframe1['B365D']
B365D_p_draw=B365D_p_draw.round(4)
dataframe1['B365D_p_draw']=B365D_p_draw


B365A_p_away=1/dataframe1['B365A']
B365A_p_away=B365A_p_away.round(4)
dataframe1['B365A_p_away']=B365A_p_away


BWH_p_home=1/dataframe1['BWH']
BWH_p_home=BWH_p_home.round(4)
dataframe1['BWH_p_home']=BWH_p_home


BWD_p_draw=1/dataframe1['BWD']
BWD_p_draw=BWD_p_draw.round(4)
dataframe1['BWD_p_draw']=BWD_p_draw


BWA_p_away=1/dataframe1['BWA']
B365A_p_away=BWA_p_away.round(4)
dataframe1['BWA_p_away']=BWA_p_away



IWH_p_home=1/dataframe1['IWH']
IWH_p_home=IWH_p_home.round(4)
dataframe1['IWH_p_home']=IWH_p_home


IWD_p_draw=1/dataframe1['IWD']
IWD_p_draw=IWD_p_draw.round(4)
dataframe1['IWD_p_draw']=IWD_p_draw


IWA_p_away=1/dataframe1['IWA']
IWA_p_away=IWA_p_away.round(4)
dataframe1['IWA_p_away']=IWA_p_away



PSH_p_home=1/dataframe1['PSH']
PSH_p_home=PSH_p_home.round(4)
dataframe1['PSH_p_home']=PSH_p_home


PSD_p_draw=1/dataframe1['PSD']
PSD_p_draw=PSD_p_draw.round(4)
dataframe1['PSD_p_draw']=PSD_p_draw


PSA_p_away=1/dataframe1['PSA']
PSA_p_away=PSA_p_away.round(4)
dataframe1['PSA_p_away']=PSA_p_away



B365total=dataframe1['B365H_p_home'] + dataframe1['B365A_p_away'] + dataframe1['B365D_p_draw']
dataframe1['B365H_p_homenormalized']=B365H_p_home/B365total
dataframe1['B365D_p_drawnormalized']=B365D_p_draw/B365total
dataframe1['B365A_p_awaynormalized']=B365A_p_away/B365total

BWtotal=dataframe1['BWH_p_home'] + dataframe1['BWA_p_away'] + dataframe1['BWD_p_draw']  
dataframe1['BWH_p_homenormalized']=BWH_p_home/BWtotal
dataframe1['BWD_p_drawnormalized']=BWD_p_draw/BWtotal
dataframe1['BWA_p_awaynormalized']=BWA_p_away/BWtotal

IWtotal=dataframe1['IWH_p_home'] + dataframe1['IWA_p_away'] + dataframe1['IWD_p_draw']
dataframe1['IWH_p_homenormalized']=IWH_p_home/IWtotal
dataframe1['IWD_p_drawnormalized']=IWD_p_draw/IWtotal
dataframe1['IWA_p_awaynormalized']=IWA_p_away/IWtotal

PStotal=dataframe1['PSH_p_home'] + dataframe1['PSA_p_away'] + dataframe1['PSD_p_draw']
dataframe1['PSH_p_homenormalized']=PSH_p_home/PStotal
dataframe1['PSD_p_drawnormalized']=PSD_p_draw/PStotal
dataframe1['PSA_p_awaynormalized']=PSA_p_away/PStotal


```

```python
B365diff=dataframe1['B365H_p_homenormalized']-dataframe1['B365A_p_awaynormalized']
BWdiff=dataframe1['BWH_p_homenormalized']-dataframe1['BWA_p_awaynormalized']
IWdiff=dataframe1['IWH_p_homenormalized']-dataframe1['IWA_p_awaynormalized']
PSdiff=dataframe1['PSH_p_homenormalized']-dataframe1['PSA_p_awaynormalized']
dataframe1['B365diff']=B365diff
dataframe1['BWdiff']=BWdiff
dataframe1['IWdiff']=IWdiff
dataframe1['PSdiff']=PSdiff



mergeallbookmakersdifference=pd.concat([dataframe1['B365diff'],dataframe1['BWdiff'],dataframe1['IWdiff'],dataframe1['PSdiff']])
mergebookmakerdraw=pd.concat([dataframe1['B365D_p_drawnormalized'],dataframe1['BWD_p_drawnormalized'],dataframe1['IWD_p_drawnormalized'],dataframe1['PSD_p_drawnormalized']])
mergeactualdraw=pd.concat([dataframe1['FTR'],dataframe1['FTR'],dataframe1['FTR'],dataframe1['FTR']])
mergeddataframe1=pd.DataFrame({'bookmakerdiff':mergeallbookmakersdifference,'bookmakerdraw':mergebookmakerdraw,'actualdraw':mergeactualdraw})

px=mergeddataframe1['bookmakerdiff']
py=mergeddataframe1['bookmakerdraw']
plt.scatter(px,py)

y, bins=np.histogram(mergeddataframe1['bookmakerdiff'],bins = [-1,-0.8,-0.6,-0.4,-0.2,0,0.2,0.4,0.6,0.8,1])
x=np.histogram(mergeddataframe1[mergeddataframe1.actualdraw=='D']['bookmakerdiff'],bins)[0]

for j in range(len(y)):   #JUST TO PREVENT ERROR OF 0/0 SITUATION I EMBED THIS FUNCTION
    if y[j]==0:
        y[j]=1
        
rng=[]
rng.append((bins[:-1] +bins[1:] )/2)
plt.scatter(rng,x/y)
plt.xlabel('homewin-awaywin')
plt.ylabel('draw')
plt.title('4 bookmakers together')


# FOR 4 BOOKMAKER WE CAN SEE ORANGE POINTS MAY BE ABOVE THE BLUE POINTS SLİGHTLY SO WE CAN SAY A LITTLE BIAS EXIST IN (0-0.2) BIN


```
