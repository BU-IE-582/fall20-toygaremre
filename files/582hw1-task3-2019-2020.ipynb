```python
import pandas as pd
import sys
import numpy as np
import matplotlib.pyplot as plt

#IN THIS PART I REMOVE THE DATA THAT INVOLVES RED CARD FOR HOME OR AWAY THEN I CREATE A NEW DATAFRAME ACCORDINGLY BELOW
#ALSO I WOULD LIKE TO EXPRESS NO SIGNIFICANT CHANGE TAKES PLACE IN PLOTS

```

```python
Matches=pd.read_csv('/Users/toygar/Desktop/IE582/E0 2019-2020.csv',sep=str,delimiter=',')
dataframe1=pd.DataFrame(Matches)

B365H_p_home=1/dataframe1[(dataframe1.HR==0) & (dataframe1.AR==0)]['B365H']
B365H_p_home=B365H_p_home.round(4)


B365D_p_draw=1/dataframe1[(dataframe1.HR==0) & (dataframe1.AR==0)]['B365D']
B365D_p_draw=B365D_p_draw.round(4)


B365A_p_away=1/dataframe1[(dataframe1.HR==0) & (dataframe1.AR==0)]['B365A']
B365A_p_away=B365A_p_away.round(4)


BWH_p_home=1/dataframe1[(dataframe1.HR==0) & (dataframe1.AR==0)]['BWH']
BWH_p_home=BWH_p_home.round(4)


BWD_p_draw=1/dataframe1[(dataframe1.HR==0) & (dataframe1.AR==0)]['BWD']
BWD_p_draw=BWD_p_draw.round(4)


BWA_p_away=1/dataframe1[(dataframe1.HR==0) & (dataframe1.AR==0)]['BWA']
B365A_p_away=BWA_p_away.round(4)



IWH_p_home=1/dataframe1[(dataframe1.HR==0) & (dataframe1.AR==0)]['IWH']
IWH_p_home=IWH_p_home.round(4)


IWD_p_draw=1/dataframe1[(dataframe1.HR==0) & (dataframe1.AR==0)]['IWD']
IWD_p_draw=IWD_p_draw.round(4)


IWA_p_away=1/dataframe1[(dataframe1.HR==0) & (dataframe1.AR==0)]['IWA']
IWA_p_away=IWA_p_away.round(4)



PSH_p_home=1/dataframe1[(dataframe1.HR==0) & (dataframe1.AR==0)]['PSH']
PSH_p_home=PSH_p_home.round(4)


PSD_p_draw=1/dataframe1[(dataframe1.HR==0) & (dataframe1.AR==0)]['PSD']
PSD_p_draw=PSD_p_draw.round(4)


PSA_p_away=1/dataframe1[(dataframe1.HR==0) & (dataframe1.AR==0)]['PSA']
PSA_p_away=PSA_p_away.round(4)

NEWdataframe1withoutredcard=pd.DataFrame({'B365H_p_home':B365H_p_home,'B365D_p_draw':B365D_p_draw,'B365A_p_away':B365A_p_away,'BWH_p_home':BWH_p_home,'BWD_p_draw':BWD_p_draw,'BWA_p_away':BWA_p_away,'IWH_p_home':IWH_p_home,'IWD_p_draw':IWD_p_draw,'IWA_p_away':IWA_p_away,'PSH_p_home':PSH_p_home,'PSD_p_draw':PSD_p_draw,'PSA_p_away':PSA_p_away,'FTR':dataframe1[(dataframe1.HR==0) & (dataframe1.AR==0)]['FTR']})



B365total=NEWdataframe1withoutredcard['B365H_p_home'] + NEWdataframe1withoutredcard['B365A_p_away'] + NEWdataframe1withoutredcard['B365D_p_draw']
NEWdataframe1withoutredcard['B365H_p_homenormalized']=B365H_p_home/B365total
NEWdataframe1withoutredcard['B365D_p_drawnormalized']=B365D_p_draw/B365total
NEWdataframe1withoutredcard['B365A_p_awaynormalized']=B365A_p_away/B365total

BWtotal=NEWdataframe1withoutredcard['BWH_p_home'] + NEWdataframe1withoutredcard['BWA_p_away'] + NEWdataframe1withoutredcard['BWD_p_draw']  
NEWdataframe1withoutredcard['BWH_p_homenormalized']=BWH_p_home/BWtotal
NEWdataframe1withoutredcard['BWD_p_drawnormalized']=BWD_p_draw/BWtotal
NEWdataframe1withoutredcard['BWA_p_awaynormalized']=BWA_p_away/BWtotal

IWtotal=NEWdataframe1withoutredcard['IWH_p_home'] + NEWdataframe1withoutredcard['IWA_p_away'] + NEWdataframe1withoutredcard['IWD_p_draw']
NEWdataframe1withoutredcard['IWH_p_homenormalized']=IWH_p_home/IWtotal
NEWdataframe1withoutredcard['IWD_p_drawnormalized']=IWD_p_draw/IWtotal
NEWdataframe1withoutredcard['IWA_p_awaynormalized']=IWA_p_away/IWtotal

PStotal=NEWdataframe1withoutredcard['PSH_p_home'] + NEWdataframe1withoutredcard['PSA_p_away'] + NEWdataframe1withoutredcard['PSD_p_draw']
NEWdataframe1withoutredcard['PSH_p_homenormalized']=PSH_p_home/PStotal
NEWdataframe1withoutredcard['PSD_p_drawnormalized']=PSD_p_draw/PStotal
NEWdataframe1withoutredcard['PSA_p_awaynormalized']=PSA_p_away/PStotal




```

```python
B365diff=NEWdataframe1withoutredcard['B365H_p_homenormalized']-NEWdataframe1withoutredcard['B365A_p_awaynormalized']
BWdiff=NEWdataframe1withoutredcard['BWH_p_homenormalized']-NEWdataframe1withoutredcard['BWA_p_awaynormalized']
IWdiff=NEWdataframe1withoutredcard['IWH_p_homenormalized']-NEWdataframe1withoutredcard['IWA_p_awaynormalized']
PSdiff=NEWdataframe1withoutredcard['PSH_p_homenormalized']-NEWdataframe1withoutredcard['PSA_p_awaynormalized']
NEWdataframe1withoutredcard['B365diff']=B365diff
NEWdataframe1withoutredcard['BWdiff']=BWdiff
NEWdataframe1withoutredcard['IWdiff']=IWdiff
NEWdataframe1withoutredcard['PSdiff']=PSdiff



mergeallbookmakersdifference=pd.concat([NEWdataframe1withoutredcard['B365diff'],NEWdataframe1withoutredcard['BWdiff'],NEWdataframe1withoutredcard['IWdiff'],NEWdataframe1withoutredcard['PSdiff']])
mergebookmakerdraw=pd.concat([NEWdataframe1withoutredcard['B365D_p_drawnormalized'],NEWdataframe1withoutredcard['BWD_p_drawnormalized'],NEWdataframe1withoutredcard['IWD_p_drawnormalized'],NEWdataframe1withoutredcard['PSD_p_drawnormalized']])
mergeactualdraw=pd.concat([NEWdataframe1withoutredcard['FTR'],NEWdataframe1withoutredcard['FTR'],NEWdataframe1withoutredcard['FTR'],NEWdataframe1withoutredcard['FTR']])
mergeddataframe1=pd.DataFrame({'bookmakerdiff':mergeallbookmakersdifference,'bookmakerdraw':mergebookmakerdraw,'actualdraw':mergeactualdraw})

px=mergeddataframe1['bookmakerdiff']
py=mergeddataframe1['bookmakerdraw']
plt.scatter(px,py)

y, bins=np.histogram(mergeddataframe1['bookmakerdiff'],bins = [-1,-0.8,-0.6,-0.4,-0.2,0,0.2,0.4,0.6,0.8,1])
x=np.histogram(mergeddataframe1[mergeddataframe1.actualdraw=='D']['bookmakerdiff'],bins)[0]

for j in range(len(y)):   #JUST TO PREVENT ERROR OF 0/0 SITUATION I EMBED THIS FUNCTION
    if y[j]==0:
        y[j]=1
        
rng=[]
rng.append((bins[:-1] +bins[1:] )/2)
plt.scatter(rng,x/y)
plt.xlabel('without red card homewin-awaywin')
plt.ylabel('without red card draw')
plt.title('without red card 4 bookmakers together')

#RELATED TO THE (0,0.2) BİN WE HAVE BIAS WİTH RESPECT TO 4 BOOKMAKERS TOGETHER


```
